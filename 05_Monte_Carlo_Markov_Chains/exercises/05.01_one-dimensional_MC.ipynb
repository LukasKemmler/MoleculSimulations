```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#
# Let's consider the following one-dimensional double-well potential,
# which may represent a real physical system, such as an ion moving along the z-direction
# in the presense of water molecules)
#
def dwp(q):
    ''' double-well potential in kJ/mol (always document units!)'''
    q2 = q - 2.0
    U = 2.0 * q2 - 6.0 * q2**2 + q2**4
    return U
a_pos = np.linspace(-0.5,4.5,50) # in nm (always document units!)
plt.figure()
plt.plot(a_pos, dwp(a_pos))
plt.title("ion surrounded water")
plt.xlabel("ion z-coordinate [nm]")
plt.ylabel("U [kJ/mol]")
```

```python
#
# Let's build our own MCMC engine (please complete the FIX MEs)
#

def mcmc_dwp(U, q, RT, step, n):
    ''' 
    one-dimensional Markov Chain Monte Carlo (MCMC) via Metropolis-Hastings:
    U    : potential energy [kJ/mol]
    q    : starting value of z-coordinate [nm]
    RT   : scaling factor for the MCMC acceptance criterion, 
           i.e. product of gas constant R (or Boltzmann constant in molar units) and temperature
           a.k.a: inverse temperature beta = 1/RT
           [kJ/mol]
    step : maximum absolute MCMC step size [nm]
    n    : number of MCMC steps
    '''
    # MCMC-sampled trajectory of z-coordinate
    a_traj  = np.zeros(shape = (n,) + np.shape(q), dtype=np.float64)
    # potential energies of MCMC-sampled trajectory
    a_epot  = np.zeros(shape = (n,), dtype=np.float64)
    # acceptances along MCMC-sampled trajectory
    a_isacc = np.zeros(shape = (n,), dtype=np.bool)
    # initial potential energy
    epot = U(q)
    for i in range(n):
        # FIX ME (start)
        # propose next q_trial = q + Δq, with Δq uniformly randomly sampled in the interval [-step,+step]
        q_trial = q + 2*step * np.random.random() - step
        # FIX ME (end)
        
        # accept q_trial? For now, set accept to False, but if accept, set to True
        accept = False
        # potential energy of q_trial
        epot_trial = U(q_trial)
        
        # FIX ME (start)
        # Use the Metropolis-Hastings algorithm to accept or reject q_trial
        # (if you get stuck here, just google "monte carlo hasting metropolis python" ;-)
        acceptance_probability = min(1.0,np.exp(-(epot_trial-epot)/RT))
        if np.random.random() < acceptance_probability:
            accept = True
        # FIX (end)

        if accept:
            q = q_trial
            epot = epot_trial

        a_isacc[i] = accept
        a_traj[i]  = q
        a_epot[i]  = epot
    return a_traj, a_epot, a_isacc
```

```python
#
# Now run the MCMC simulation, after having checked the following parameters
#
# initial z-coordinate
q_init = 0
# inverse temperature beta = 1/RT
RT = 0.6160340502099109 # Boltzmann constant [kcal/mol]
# maximum absolute MCMC step size [nm]
step = 0.5
# number of MCMC steps
n = 100000

a_traj, a_epot, a_isacc = mcmc_dwp(dwp, q_init, RT, step, n)
```

```python
# TASK:
# use the a_isacc array to compute the acceptance rate (ar) of your MCMC run
acc_rate = np.mean(a_isacc)
print acc_rate

#
# FYI:
# for a one-dimensional (N=1) Gaussian distribution, the ideal value of ar is 50%, for higher N, ar_ideal -> 23%
# if ar is too low (high), the MCMC chain will mix and converge slowly, try decreasing (increasing) the step size
#
# TASK:
# Please explain how 
# a) a too low (high) step size may result in a too high (low) acceptance rate, and how
# b) a too high (low) acceptance rate may negatively affect the convergance of an MCMC chain
# ANSWERS:
# a) A low step size results in a  small change of energy. Thus the energy difference is small and the acceptance 
#    propability is high which gives a high acceptance rate.
#    For a high step size the change in energy is large and therefore the acceptance probability is low. Thos leads
#    to a small acceptance rate.
# b) When the acceptance rate is small the algorithem stays almost at the same position. With a large acceptance
#    rate the point moves nearly independent of the underlying potential.
#
```

```python
#
# Let's compare the sampled potential energy with the sampled free energy G, which relates 
# to probabilities via
# G(q) = - C * RT * ln (Probability(q)), where C is a normalization constant
#

numbins = 200
# use gauge invariance to set the minimum potential to zero, and plot
# sampled potential values vs. z-coordinate (blue)
plt.scatter(a_traj, a_epot - min(a_epot), label = "U (sampled) [kJ/mol]")

# compute a distribution density of the visited z-coordinates by making a histogram
startbin = -1
endbin   =  4
myhist = np.histogram(a_traj, bins = numbins, range = (startbin, endbin), density = True)

# This distribution density in fact represents a Boltzmann distribution
# because it was generated with MCMC, which fulfills detailed balance (see lecture notes).
# Therefore, we can compute a free energy profile from this distribution
# (normalizing for binsizes != 1):
a_mcmc_G = - RT * np.log( 1. * myhist[0] * (endbin - startbin) / numbins)
a_mcmc_G -= min(a_mcmc_G)

# This free energy profile can be compared to the potential energy computed above
# for the one-dimensional case, in fact, the free energy then should be the same as the potential energy
# since every z-coordinate "state" is realized by exactly one structurual arrangement (no entropy)
plt.scatter(myhist[1][:-1], a_mcmc_G, label = "G [kJ/mol]")
plt.plot(a_pos, dwp(a_pos) - min(dwp(a_pos)), label = "U (full) [kJ/mol]", color = "k")
plt.xlabel("ion z-coordinate [nm]")
plt.legend()
```
